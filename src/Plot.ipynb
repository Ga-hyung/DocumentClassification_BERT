```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
with open('K-Accuracy.pickle', 'rb') as f:
    accuracy = pickle.load(f)
```

```python
with open('Majority_accuracy.pickle','rb') as f:
    majority_acc = pickle.load(f)
```

```python
accuracy
```

```python
majority_acc
```

```python
x = [1,2,3,4,5]
values = range(len(x))
plt.plot(values, accuracy, marker = 'o', color = 'orange', label = 'BERT')
plt.plot(values, majority_acc, marker = 'o', color = 'Grey', label = 'Majority of Class')
plt.xticks(values, x)
plt.xlabel("K-Fold")
plt.ylabel("Accuracy")
plt.legend()
plt.title("K-Fold Cross Validation")
```
